### 1. Disparate Scales
**Outcome**: Get a biased models, inaccurate results, and unstable parameter estimates

**Solutions**: Standardization

```python
# import
import pandas as pd
import numpy as np
```

```python
# Create a sample data set in disparate scale
sample_df = pd.DataFrame({'v1': pd.Series(np.random.choice(1000,20)),
                          'v2': pd.Series(np.random.choice(20,20))})
sample_df.head()
```

```python
# Standardize
sample_df_std = sample_df.copy()  #In order to run this cell many times w/o error

# Apply z-score formula to new column
for col_name in sample_df.columns:
    new_col = col_name + '_std'
    sample_df_std[new_col] = (sample_df[col_name] - sample_df[col_name].mean()) / sample_df[col_name].std()

sample_df_std.head()
```

### 2. Noise & Outliers
**Solutions 1**: Discretization

- Changing a numeric feature into an ordinal or nominal categorical features based on value ranges, also referred to as binning.
- Commonly used in linear models to increase flexibility.
- Smoothes complex signals in training data. **Decrease overfitting**.
- Deals with [missing values](#3.-Missing-Values:-Uncollected-Information) or outliers.

```python
# import
import pandas as pd
import numpy as np
```

```python
# Create a sample data set
sample_df = pd.DataFrame({'v1': pd.Series(np.random.randn(20))})

sample_df.head()
```

```python
# Discretize (Creating Bins)
sample_df['v1_discret'] = pd.DataFrame(pd.cut(sample_df['v1'], 5))

sample_df.head(8)
```

**Solution 2**: Winsorizing

- Removing outliers in a feature's value and replacing them with more central values of that feature.

```python
# Import
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
```

```python
# Create a sample data set
sample_df = pd.DataFrame({'v1': pd.Series(np.random.choice(1000,20))})

sample_df
```

```python
# Winsorize
# Replace the min & Max 10% of data
sample_df['v1_winsorized'] = winsorize(sample_df['v1'], limits = [0.1, 0.1])
sample_df
```

### 3. Missing Values: Uncollected Information
**Solutions 1**: Imputation

- Replacing missing data with an appropriate, non-missing value.
- **Be cautious when using imputation in predictive modeling**. Missingness is oftern predictive. Imputation changes the distribution of the input feature learned by the model.

```python
# Import
import numpy as np
import pandas as pd
```

```python
# Create sample data set
sample_df = pd.DataFrame({'v1': [0,1,2,3, np.nan, 5,6,7, np.nan, 8,9]})

sample_df.head()
```

```python
sample_df['v1_impute'] = sample_df.fillna(sample_df['v1'].mean())

sample_df.head()
```

### 4. Character Features: Encoding
**Solution 1**: Simple Encoding

- Changing the representation of a feature. Encoding categorical character features to numeric features.

```python
# Import
import pandas as pd

#Create a sample data set
sample_df = pd.DataFrame({'x1': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
                          'x2': ['C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']})
sample_df
```

```python
# Encode categorical variables as binary dummies
pd.concat([sample_df, pd.get_dummies(sample_df)], axis = 1)

# concat: merge dataframe together
# axis = 1: combine by rows
```
